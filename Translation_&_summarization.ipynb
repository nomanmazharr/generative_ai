```python
from transformers import pipeline
```

```python
translate= pipeline(
    'translation',
    model = 'facebook/nllb-200-distilled-600M'
)
```

```python
text = '''
Hey, How you doing?
'''
```

```python
translate(text,
          src_lang = 'eng_latn',
          tgt_lang = 'urd_Arab')
```

```python
text1 = '''
  A connection to the notebook server could not be established.
  The notebook will continue trying to reconnect. Check your network connection or notebook server configuration.
'''
```

```python
translate(text1,
          src_lang = 'eng_Latn',
          tgt_lang = 'urd_Arab')
```
