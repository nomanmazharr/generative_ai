```python
import os
import gradio as gr
```

```python
app = gr.Blocks()
```

### Model to transcribe the speech

```python
def transcribe(file):
  if file is None:
    gr.warning('No file found')
    return ""
  else:
    # output = speech_converter(file)
    output = any_language_converter(file)

    return output['text']
```

```python
file_transcribe = gr.Interface(
    fn = transcribe,
    inputs = gr.Audio(
        sources = 'upload',
        type = 'filepath'

    ),
    outputs = gr.Textbox(
        lines=3,
        label = 'Transcription'
    )
)
```

```python
mic_transcribe = gr.Interface(
    fn = transcribe,
    inputs = gr.Audio(
        sources = 'microphone',
        type = 'filepath'
    ),
    outputs = gr.Textbox(
        label = 'Transcription',
        lines=5
    )

)
```

```python
with app:
  gr.TabbedInterface(
      [file_transcribe, mic_transcribe]
  )
```

```python
app.launch(share=True)
```

```python
app.close()
```
