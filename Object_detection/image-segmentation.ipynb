```python
from transformers import pipeline
```

```python
segmentation = pipeline(
    'image-segmentation',
    model = 'briaai/RMBG-1.4',
    trust_remote_code ='True'
    )
```

### Importing gradio and defining function for segmentation

```python
import gradio as gr
```

```python
def segment_image(img):
    segmented_image = segmentation(img)
    return segmented_image
```

### Interface

```python
app = gr.Interface(
    fn = segment_image,
    inputs = gr.Image(label='input', type='pil'),
    outputs = gr.Image(label='output', type='pil')
)
```

```python
app.launch(share=True)
```

```python
app.close()
```
